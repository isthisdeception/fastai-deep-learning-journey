```python
# frist training loop
weights = torch.randn(784, requires_grad = True)
bias = torch.randn(1, requires_grad=True)
label = tensor(1.)
lr = 0.0001
for i in range(20):
    prediction = (flat_img * weights).sum() + bias
    loss = (prediction - label)**2
    loss.backward()
    with torch.no_grad():
        weights -= lr * weights.grad
        bias -= lr * bias.grad

    weights.grad.zero_() # erase old gradients before calculating new gradients
    bias.grad.zero_()
    print(f"Step {i:2d} | Loss = {loss.item():.4f}")
```

**So far we have done all the stuff for one image, now it's time to do it for multiple image.**
A real model learns from many examples

```python
# Load several image
threes = (path/'train'/'3').ls()[:20]
sevens = (path/'train'/'3').ls()[:20]

len(threes), len(sevens)
```

```python
# Create a dataset
xs = []
ys = []

for p in threes:
    img = PILImage.create(p).convert('L')
    x = tensor(img).float()/255
    x = x.view(-1)

    xs.append(x)
    ys.append(1.)

for p in sevens:
    img = PILImage.create(p).convert('L')
    x = tensor(img).float()/255
    x = x.view(-1)

    xs.append(x)
    ys.append(0.)
```

```python
# stack everything
x_train = torch.stack(xs)
y_train = tensor(ys)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
weights = torch.randn(784, requires_grad = True)
bias = torch.randn(1, requires_grad = True)
prediction = x_train @ weights + bias
prediction.shape
```

```python
prediction[:10]
```

```python
# calculating the loss
loss = ((prediction - y_train)**2).mean()
loss
```

**It's time to train on all 40 Image**

```python
weights = torch.randn(784, requires_grad = True)
bias = torch.randn(1, requires_grad = True)

lr = 0.0001
for i in range(20):
    prediction = x_train @ weights + bias
    loss = ((prediction - y_train)**2).mean() # since we are doing it for 40 image

    loss.backward()

    with torch.no_grad():
        weights -= lr * weights.grad
        bias -= lr * bias.grad

    weights.grad.zero_()
    bias.grad.zero_()
    print(f"Step {i:2d} | Loss = {loss.item():.4f}")
```

**Entire training Loop in Plain English.**
1. Use the current weights and biases to make predictions
2. Compare predictions with the correct labels
3. Calculate the loss (how wrong we are)
4. Run backward() to calculate gradients
5. Update weights and biases using the gradients
6. Clear gradients so they don't accumulate
7. Record/print the loss
8. Repeat for the next epoch

```python
# code 
class myLearner:  
    def __init__(self, x_train, y_train, weights1, bias1, weights2, bias2, lr):
        self.x_train = x_train
        self.y_train = y_train
        self.weights1 = weights1
        self.bias1 = bias1
        self.weights2 = weights2
        self.bias2 = bias2
        self.lr = lr

    def fit(self, epochs):
        for i in range(epochs):
            hidden = self.x_train @ self.weights1 + self.bias1

            hidden = hidden.clamp(min=0)
        
            output = hidden @ self.weights2 + self.bias2
        
            loss = ((output - self.y_train)**2).mean()
        
            loss.backward()
        
            with torch.no_grad():
                self.weights1 -= self.lr * self.weights1.grad
                self.bias1 -= self.lr * self.bias1.grad
        
                self.weights2 -= self.lr * self.weights2.grad
                self.bias2 -= self.lr * self.bias2.grad
        
            self.weights1.grad.zero_()
            self.bias1.grad.zero_()
        
            self.weights2.grad.zero_()
            self.bias2.grad.zero_()
        
            print(f"Step {i:2d} | Loss = {loss.item():.4f}")

    def predict(self, x):
        with torch.no_grad():
            hidden = x @ self.weights1 + self.bias1
            hidden = hidden.clamp(min=0)
            output = hidden @ self.weights2 + self.bias2
            output = output.item()
        if output > 0.5:
            return 3
        else:
            return 7
        
learn = myLearner(
    x_train, 
    y_train, 
    weights1, 
    bias1, 
    weights2, 
    bias2, 
    lr
)

learn.fit(20)
# x should be flattened if it's a 2D image, e.g. (28x28 → 784
x = image.flatten().float()/255  
pred = learn.predict(x)
```
